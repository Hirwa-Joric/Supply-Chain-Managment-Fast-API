```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from faker import Faker
import random

def generate_dummy_data(num_records=1000):
    fake = Faker()
    
    # Set random seed for reproducibility
    np.random.seed(42)
    random.seed(42)
    
    # Generate data
    data = {
        'transaction_id': [f'TRX{str(i).zfill(6)}' for i in range(1, num_records + 1)],
        'product_id': [f'PRD{str(random.randint(1, 100)).zfill(3)}' for _ in range(num_records)],
        'product_name': [fake.word() + ' ' + random.choice(['Tool', 'Part', 'Equipment', 'Supply']) for _ in range(num_records)],
        'category': [random.choice(['Tools', 'Hardware', 'Safety', 'Electronics', 'Machinery']) for _ in range(num_records)],
        'quantity': np.random.randint(1, 100, num_records),
        'unit_price': np.random.uniform(10, 1000, num_records).round(2),
        'supplier': [fake.company() for _ in range(num_records)],
        'transaction_type': [random.choice(['PURCHASE', 'SALE', 'RETURN', 'ADJUSTMENT']) for _ in range(num_records)],
        'transaction_date': [(datetime.now() - timedelta(days=random.randint(0, 365))).strftime('%Y-%m-%d') for _ in range(num_records)],
        'is_active': [random.choice([True, True, True, False]) for _ in range(num_records)],  # Bias towards True
    }
    
    # Calculate total_value
    data['total_value'] = (np.array(data['quantity']) * np.array(data['unit_price'])).round(2)
    
    # Create DataFrame
    df = pd.DataFrame(data)
    
    # Sort by transaction_date
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])
    df = df.sort_values('transaction_date')
    
    # Reset index
    df = df.reset_index(drop=True)
    
    return df

def save_to_csv(df, filename='inventory_transactions.csv'):
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")

def main():
    # Generate data
    df = generate_dummy_data(1000)
    
    # Save to CSV
    save_to_csv(df)
    
    # Print first few rows and basic statistics
    print("\nFirst few rows of the generated data:")
    print(df.head())
    print("\nBasic statistics:")
    print(df.describe())
    print("\nData types:")
    print(df.dtypes)

if __name__ == "__main__":
    main()
```
